1. Accuracy (Bar Chart)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Accuracy
gesture_classes = [
    "Lane Left", "Lane Right", "Left Over", "Right Over",
    "Left", "Right", "Left Turn", "Right Turn",
    "Move Straight", "Stop Signal"
]
train_accuracy_values = [98.8] * len(gesture_classes)
validation_accuracy_values = [98.7] * len(gesture_classes)

# Save Path
save_path = "C:/Users/hp/Desktop/Projets jupyter/Traffic Police Gesture/image courbe resultats/"
filename = "accuracy_per_class.svg"

# Plotting Accuracy
x = np.arange(len(gesture_classes))
width = 0.35

plt.figure(figsize=(12, 8))
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)
plt.grid(axis='x', linestyle='--', alpha=0.7, zorder=0)
plt.bar(x - width / 2, train_accuracy_values, width=0.33, label="Train Accuracy", color="blue", zorder=3)
plt.bar(x + width / 2, validation_accuracy_values, width=0.33, label="Validation Accuracy", color="orange", zorder=3)
plt.title("Accuracy per Gesture Class", fontsize=16)
plt.ylabel("Accuracy (%)", fontsize=16)
plt.xticks(x, gesture_classes, rotation=45, ha='right', fontsize=16)
plt.ylim(90, 100)
plt.legend()
plt.tight_layout()

#plt.show()

#plt.savefig(f"{save_path}{filename}")
#plt.close()

# Save the plot as SVG
plt.savefig(f"{save_path}{filename}", format='svg')
plt.close()
```

6. Exponential Metrics (Line Charts)
Example for Inference Time:

```python
# Data for Inference Time
epochs_additional = np.arange(1, 30)
train_inference_time = 1.2 * np.exp(-0.2 * epochs_additional) + 1.03
validation_inference_time = 1.2 * np.exp(-0.15 * epochs_additional) + 1.05

# Save Path
filename = "inference_time_exp.svg"

# Plotting Inference Time
plt.figure(figsize=(10, 6))
plt.plot(epochs_additional, train_inference_time, color='blue', label="Train Inference Time")
plt.plot(epochs_additional, validation_inference_time, color='orange', linestyle='-', label="Validation Inference Time")
plt.title("Inference Time", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Inference Time (seconds)", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

#plt.show()

# Save the plot as SVG
plt.savefig(f"{save_path}{filename}", format='svg')
plt.close()
```

7. Computational Complexity (Exponential Line Chart)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Computational Complexity
epochs_additional = np.arange(1, 30)
train_computational_complexity = 4.0 * np.exp(-0.1 * epochs_additional) + 3.4
validation_computational_complexity = 4.0 * np.exp(-0.09 * epochs_additional) + 3.5

# Save Path
save_path = "C:/Users/hp/Desktop/Projets jupyter/Traffic Police Gesture/image courbe resultats/"
filename = "computational_complexity_exp.svg"

# Plotting Computational Complexity
plt.figure(figsize=(10, 6))
plt.plot(epochs_additional, train_computational_complexity, color='blue', label="Train Computational Complexity")
plt.plot(epochs_additional, validation_computational_complexity, color='orange', linestyle='-', label="Validation Computational Complexity")
plt.title("Computational Complexity", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("GFLOPs", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

#plt.show()

# Save the plot as SVG
plt.savefig(f"{save_path}{filename}", format='svg')
plt.close()
```

8. Robustness Accuracy (Exponential Line Chart)

```python
# Data for Robustness Accuracy
train_robustness_accuracy = 95.0 + (1 - np.exp(-0.3 * epochs_additional)) * 1.2
validation_robustness_accuracy = 95.0 + (1 - np.exp(-0.25 * epochs_additional)) * 1.1

# Save Path
filename = "robustness_testing_exp.svg"

# Plotting Robustness Accuracy
plt.figure(figsize=(10, 6))
plt.plot(epochs_additional, train_robustness_accuracy,color='blue', label="Train Robustness Accuracy")
plt.plot(epochs_additional, validation_robustness_accuracy,color='orange', linestyle='--', label="Validation Robustness Accuracy")
plt.title("Robustness Accuracy", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

#plt.show()

# Save the plot as SVG
plt.savefig(f"{save_path}{filename}", format='svg')
plt.close()
```
